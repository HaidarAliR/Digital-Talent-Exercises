```python
#Report the account representative for each customer.
run("""
SELECT
    c.customerName,
    c.salesRepEmployeeNumber,
    CONCAT (e.firstName,' ',e.lastName) AS employeeName
FROM
    customers c
    LEFT JOIN
    employees e ON c.SalesRepEmployeeNumber = e.employeeNumber
ORDER BY customerName
""")
```

```python
#Report total payments for Atelier graphique. (INNER JOIN customers with orders)
run("""
SELECT
    customerName,
    customerNumber,
    SUM(p.amount)
FROM customers
    INNER JOIN
    payments p USING (customerNumber)
WHERE customerName = 'Atelier graphique'
    """)
```

```python
#Report the total payments by date
run("""
SELECT
    paymentDate,
    SUM(amount)
FROM
    payments
GROUP BY
    paymentDate
ORDER BY
    paymentDate
    """)
```

```python
# Report the products that have not been sold. (INNER JOIN products and orderdetails)
run("""
SELECT
    productCode,
    productName
FROM
    products
WHERE
    productCode NOT IN (SELECT productCode FROM orderdetails)
    """)
```

```python
# List the amount paid by each customer. (INNER JOIN customer and payment)
run("""
SELECT
    c.customerName,
    SUM(p.amount)
FROM 
    customers c
INNER JOIN
    payments p USING (customerNumber)
GROUP BY
    customerName
""")
```

```python
#How many orders have been placed by Herkku Gifts?
run("""
SELECT
    customerName,
    COUNT(customerName) AS '#orders'
FROM
    customers
INNER JOIN
    payments USING (customerNumber)
WHERE
    customerName = 'Herkku gifts'
    """)
```

```python
# Who are the employees in Boston?
run("""
SELECT
    firstName AS 'EmployeeInBoston',
    city
FROM
    employees
INNER JOIN
    offices USING (officeCode)
WHERE
    city = 'Boston'
    """)
```

```python
# Report those payments greater than $100,000. Sort the report so 
# the customer who made the highest payment appears first.
run("""
SELECT
    customerName,
    amount
FROM
    customers
INNER JOIN
    payments USING (customerNumber)
WHERE
    amount > 100000
ORDER BY
    amount DESC
    """)
```

```python
# List the value of 'On Hold' orders.
run("""
SELECT
    customerNumber,
    status,
    amount,
    orderNumber
FROM
    orders
INNER JOIN
    payments USING (customerNumber)
WHERE
    status = 'On hold'
    """)
```

```python
# Report the number of orders 'On Hold' for each customer.
run("""
SELECT
    customerNumber,
    status,
    COUNT(status) AS '#OnHold',
    customerName
FROM
    orders
INNER JOIN
    payments USING (customerNumber)
INNER JOIN
    customers USING (customerNumber)
WHERE
    status = 'On hold'
GROUP BY
    customerNumber
    """)
```

```python
# List products sold by order date.
run("""
SELECT
    productName,
    orderDate
FROM 
    products
INNER JOIN
    orderdetails USING (productCode)
INNER JOIN
    orders USING (orderNumber)
GROUP BY
    orderDate
    """)
```

```python
# List the order dates in descending order for orders for the 1940 Ford Pickup Truck.
run("""
SELECT
    productName,
    orderDate
FROM 
    products
INNER JOIN
    orderdetails USING (productCode)
INNER JOIN
    orders USING (orderNumber)
WHERE
    productName = '1940 Ford Pickup Truck'
ORDER BY
    orderDate DESC
    """)
```

```python
# List the names of customers and their corresponding 
# order number where a particular order from that customer has a value greater than $25,000?
# run("""
# SELECT
#     customerName,
#     orderNumber,
#     amount
# FROM
#     customers
# INNER JOIN
#     orders USING (customerNumber)
# INNER JOIN
#     payments USING (customerNumber)
# GROUP BY
#     customerName
# HAVING
#     amount > 25000
# """)

run("""
SELECT
    customerName,
    orderNumber,
    SUM(priceEach * quantityOrdered) AS 'total'
FROM
    customers
INNER JOIN
    orders USING (customerNumber)
INNER JOIN
    orderdetails USING (orderNumber)
GROUP BY
    customerName
HAVING
    SUM(priceEach * quantityOrdered) > 25000
    """)
```

```python
# Are there any products that appear on all orders?

run("""
SELECT
    orderNumber
FROM
    products
ORDER BY
    productCode
    """)
```

```python
# List the names of products sold at less than 80% of the MSRP.
run("""
SELECT
    priceEach,
    MSRP,
    productName
FROM
    orderdetails
INNER JOIN
    products USING (productCode)
WHERE
    priceEach < 80/100*(MSRP)
GROUP BY
    """)
```

```python
# Reports those products that have been sold with a markup 
# of 100% or more (i.e., the priceEach is at least twice the buyPrice)

run("""
SELECT
    productName,
    buyPrice,
    priceEach
FROM
    products
INNER JOIN
    orderDetails USING (productcode)
WHERE
    priceEach > 2*buyPrice
    """)
```

```python
# What is the quantity on hand for 
# products listed on 'On Hold' orders?
run("""
SELECT
    quantityInStock,
    status,
    productName
FROM
    products
INNER JOIN
    orderdetails USING (productCode)
INNER JOIN
    orders USING (orderNumber)
WHERE
    status = 'On Hold'
    """)
```

```python
# List the products ordered on a Monday.
run("""
SELECT
    productName,
    DAYNAME(orderDate) AS 'DAY'
FROM
    products
INNER JOIN
    orderdetails USING (productCode)
INNER JOIN
    orders USING (orderNumber)
WHERE
    DAYNAME(orderDate) = 'Monday'
GROUP BY
    productName
    """)
```
